# 从Posts.xml中批量获取API全部SO数据集

```python
from bs4 import BeautifulSoup
import html, re, time
```

```python
#API名称批量加入
APInamesDic = {}
```

```python
APInamesDic[1] = ['Rackspace','cloud']
APInamesDic
```

```python
APInamesDic
```

```python
APInamesDic[53] = ['Amazon','SNS']
APInamesDic
```

```python
load_path = 'stackoverflow.com-Posts/Posts.xml'
save_path = 'API_data2/'
```

```python
#每个API数据保存为一个txt文件
names = []
for key in APInamesDic:
    name = ''
    for ele in APInamesDic[key]:
        name += ele
    print(name)
    names.append(name)
    f = open(save_path+name+'.txt', 'w', encoding='utf-8')
    f.close()
names
```

```python
#关键词匹配和标签匹配结合
cnt = 0
timestart = time.time()
with open (load_path, 'r', encoding='utf-8') as f:
    for line in f:
        cnt += 1
        if cnt%1000 == 0:
            timenow = time.time()
            print('cnt:', cnt, '\ttime:', timenow-timestart)
        soup = BeautifulSoup(line, 'lxml')
        row = soup.find('row')
        try:
            tags = row.get('tags')
            title = row.get('title')
            body = row.get('body')
            for key in APInamesDic:
                flag1 = True
                flag2 = True
                flag3 = True
                for word in APInamesDic[key]:
                    if not re.search(word, tags, re.IGNORECASE):
                        flag1 = False
                    if not re.search(word, title, re.IGNORECASE):
                        flag2 = False
                    if not re.search(word, body, re.IGNORECASE):
                        flag3 = False
                if flag1 or flag2 or flag3:
                    #print(cnt, key, names[key])
                    f = open(save_path+names[key]+'.txt', 'a', encoding='utf-8')
                    f.write(line)
                    f.close()
        except:
            pass
```
